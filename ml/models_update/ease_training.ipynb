```python
# data
ratings_with_internal_id = pd.read_csv('../../maintenance/data/ratings_with_internal_id.csv')
```

```python
# data transformations
data = ratings_with_internal_id['rating'].values
row_indices = ratings_with_internal_id['user_internal_id'].values
col_indices = ratings_with_internal_id['internal_id'].values

num_users = ratings_with_internal_id['user_internal_id'].max() + 1
num_movies = ratings_with_internal_id['internal_id'].max() + 1

user_item_matrix = csr_matrix((data, (row_indices, col_indices)), shape=(num_users, num_movies))
```

```python
# model training
ease_model = EASE(reg_weight=925.0)
ease_model.fit(user_item_matrix)
```

```python
# saving model
ease_model.save_model()
```
